
## Inheritence 
Inheritance allows us to define a class that inherits all the methods and properties from another class.












```python
# Single Inheritence 

class Product:
    def __init__(self,name, price, discount):
        self.name = name 
        self.price = price 
        self.discount = discount 
    def product_Description(self):
        print(f"Product Name : {self.name}\nPrice : {self.price}\nDiscount : {self.discount}")

    def displayFinalPrice(self):
        return self.price * (1 - (self.discount / 100))

class LightProduct(Product):
    def __init__(self,name, price, discount, weight):
        super().__init__(name, price, discount) 
        self.weight = weight 
    
    def LightProduct_Description(self):
        print(f"Product Name : {self.name}\nPrice : {self.price}\nDiscount : {self.discount}\nProduct Weight : {self.weight}")

    def applyFinalDiscount(self):
        if self.weight <= 5:
            self.discount += 5 
        

p1 = Product('Rice', 1000, 10)
p2 = LightProduct('Silk',400, 10, 0.450) 
```

```python
p1.product_Description() 
print(f"Final Price : {p1.displayFinalPrice()}") 
```

```python
p2.LightProduct_Description() 
```

```python
print(f"Normal Discount : {p2.displayFinalPrice()}")
p2.applyFinalDiscount()
print(f"Additional Discount : {p2.displayFinalPrice()}")
 
```

```python
class User:
    def __init__(self, email: str, phone: int, password: str) -> None:
        self.email = email
        self.phone = phone
        self.password = password

    def browsing(self):
        print(f"Welcome To Amazon! Flat 10% Discount on Mobiles....")

class Customer(User):
    def __init__(self, email: str, phone: int, password: str, product_name: str, price: float, discount: float, address: str, payment_type: str):
        super().__init__(email, phone, password)
        self.product_name = product_name
        self.price = price
        self.discount = discount
        self.address = address
        self.payment_type = payment_type

    def applyDiscountCoupons(self):
        if self.price >= 2000:
            self.discount = 10
        elif self.price < 2000 and self.price >= 1000:
            self.discount = 5
        else:
            self.discount = 3

    def getFinalPrice(self):
        return self.price * (1 - (self.discount / 100))

class PremiumCustomer(Customer):
    def __init__(self, email: str, phone: int, password: str, product_name: str, price: float, discount: float, address: str, payment_type: str, premiumAccountId: str, premiumDeliveryTime: str, premiumPlan_InMonths: int):
        super().__init__(email, phone, password, product_name, price, discount, address, payment_type)
        self.premiumAccountId = premiumAccountId
        self.premiumDeliveryTime = premiumDeliveryTime
        self.premiumPlan_InMonths = premiumPlan_InMonths

    def applyPremiumCoupons(self):
        # Apply premium discount logic
        if self.price >= 2000:
            self.discount = 20
        elif self.price < 2000 and self.price >= 1000:
            self.discount = 10
        else:
            self.discount = 5
```

```python
p1 = User('JackSparrow', 2398472983, 'Jack@Sparrow_Pirate#1')
p2 = Customer('JackSparrow', 2398472983, 'Jack@Sparrow_Pirate#1', 'The Black Pearl', 100000, 5, 'Carribean Sea', 'loot cash') 
p3 = PremiumCustomer('JackSparrow', 2398472983, 'Jack@Sparrow_Pirate#1', 'The Black Pearl', 100000, 5, 'Carribean Sea', 'loot cash', 'azp001', 1, 12) 
```

```python
p1.browsing() 
```

```python
print(f"Initial Price : {p2.getFinalPrice()}")
print(f"After Applying discount ") 
p2.applyDiscountCoupons() 
print(f"Final Price : {p2.getFinalPrice()}")
```

```python
print(f"Premium Customer") 
print(f"Initial Price : {p3.getFinalPrice()}")
print(f"After Applying discount ") 
p3.applyPremiumCoupons() 
print(f"Final Price : {p3.getFinalPrice()}")
```

```python
class Phone:
    def make_call(self):
        return "Making Calls" 

    def send_sms(self):
        return "Sending SMS"

class Computer:
    def browsing(self):
        return "Browsing Over Internet" 
    
    def send_email(self):
        return "Sending Email" 

class SmartPhone(Phone, Computer):
    def __init__(self, phone_brand, phone_model):
        self.brand = phone_brand
        self.model = phone_model

    def printDetails(self):
        print(f"Phone : {self.brand} {self.model}")
```

```python
p = SmartPhone("Iphone", "16 Pro Max")
```

```python
p.printDetails() 
print(p.make_call()) 
print(p.send_sms()) 
print(p.browsing()) 
print(p.send_email()) 
```

```python
class Vehicle:
    def brake(self):
        print(f"Vehicle is breaking") 
    
    def accelerate(self):
        print(f"vehicle is Accelerating") 

    def left_turn(self):
        print(f"Vehicle is Turning Left") 
    
    def right_turn(self):
        print(f"Vehicle is Turning Right") 

class Car(Vehicle):
    def Enable_MoonRoof(self):
        print(f"MoonRoof Enabled") 
    
    def Telivision(self):
        print(f"Movie Time In Car") 
    
class Aeroplane(Vehicle):
    def TakeOff(self):
        print(f"Flight is Taking off") 
    
    def Landing(self):       
        print(f"Flight Has Successfully Landed") 
    
class Cruise(Vehicle):
    def Depart(self):
        print(f"Cruise has Departed") 
            
    def fishing(self):
        print(f"Cruise is Floating Still and Everyone is enjoying fishing") 

```

```python
class BankAccount:
    def __init__(self, balance):
        self.balance = balance 
    
    def deposit(self, ammount):
        self.balance += ammount 

    def withdraw(self, ammount):
        if self.balance < ammount:
            print(f"Insufficient Balance!") 
        else:
            self.balance -= ammount 

class SavingAccount(BankAccount):
    def __init__(self, balance, interest_rate):
        BankAccount.__init__(self, balance) 
        self.interest_rate = interest_rate
    
    def calculateInterest(self):
        return self.balance * self.interest_rate/100

class CreditCard(BankAccount):
    def __init__(self, balance, credit_limit):
        BankAccount.__init__(self, balance) 
        self.credit_limit = credit_limit 

    def makePurchase(self, ammount):
        if self.balance + self.credit_limit >= ammount:
            self.balance -= ammount
        
            if self.balance < 0:
                self.credit_limit += self.balance 
                self.balance = 0 

        else:
            print(f"Credit Limit Exceeded") 

class StudentAccount(SavingAccount, CreditCard):
    def __init__(self, balance, interest_rate, credit_limit, student_discount):
        SavingAccount.__init__(self, balance, interest_rate) 
        CreditCard.__init__(self, balance, credit_limit) 
        self.student_discount = student_discount
    
    def applyDiscount(self, price):
        return price * (1 - (self.student_discount/100)) 
```

```python
sa = StudentAccount(10000, 5, 5000, 10)
```

```python
square = lambda x : x ** 2 
cube = lambda x : x ** 3 

def isPrime(n : int) -> bool:
    if n <= 1:
        return False 

    for i in range(2, n):
        if n % i == 0:
            return False 

    else:
        return True 

def getSquarePrime(fun1, fun2, var):
    sq = fun1(var) 

    li = [] 
    for i in range(1, sq+1):
        if fun2(i):
            li.append(i) 

    return li  

print(getSquarePrime(cube, isPrime, int(input())))
```
